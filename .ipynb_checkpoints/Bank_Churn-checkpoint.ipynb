```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
```

```python
df = pd.read_csv('Bank_Churn_Messy.csv')
```

```python
df1 = pd.read_csv('Bank_Churn_Messy 1.csv')
```

```python
bank = pd.merge(df,df1,on = 'CustomerId')
```

```python
# Converting Estimated Salary and balance  column in correct format.
bank['EstimatedSalary'] = bank['EstimatedSalary'].replace('[€]','',regex=True).astype(float)
bank['Balance'] = bank['Balance'].replace('[€]','',regex=True).astype(float)
```

```python
bank_2 = bank1.copy()
bank_2['Geography'].value_counts()
```

```python
bank_2['Age'] = bank_2['Age'].fillna(bank_2['Age'].sum()/len(bank_2['Age']))
bank_2.isnull().sum()
```

```python
bank1.isnull().sum()
```

```python
X = bank_2.drop(columns=['Exited','Tenure_y'])
Y = bank_2['Exited']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train['Geography'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[('tnf1',OneHotEncoder(drop='first',sparse=False),['Geography','Gender','HasCrCard','IsActiveMember']),
                                              ('tnf2',MinMaxScaler(),['EstimatedSalary','Age','CreditScore','Balance'])],remainder='passthrough')
```

```python
x_train_transformed = transformer.fit_transform(x_train)
x_test_transformed = transformer.transform(x_test)
```

```python
x_test_transformed.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
Rf = RandomForestClassifier(max_depth=9,bootstrap=True,n_estimators=200,max_samples=0.75)
Rf.fit(x_train_transformed,y_train)
```

```python
y_pred=Rf.predict(x_test_transformed)
y_pred_train=Rf.predict(x_train_transformed)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print('Test accuracy',accuracy_score(y_pred,y_test))
print('Training accuracy',accuracy_score(y_pred_train,y_train))
```

```python
print(metrics.classification_report(y_train,y_pred_train))
print('**'*10)
print(metrics.classification_report(y_test,y_pred))
```

```python
print(metrics.confusion_matrix(y_train,y_pred_train))
print('**'*10)
print(metrics.confusion_matrix(y_test,y_pred))
```

```python
# Detecting Outliers 
numerical_features = ['CreditScore','Age','Tenure_x','EstimatedSalary','Balance']

for i in numerical_features:
    sns.boxplot(bank1[i])
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Density')
    plt.show()
# From boxplot we can see that the creditscore column and age column has outliers
```

```python
bank1.head()
```

```python
# Checking Distribution of numerical features.
numerical_features = ['CreditScore','Age','Tenure_x','EstimatedSalary','Balance']

for i in numerical_features:
    sns.kdeplot(bank1[i])
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Density')
    plt.show()
# Here we can see thar Credit score is normally distributed and age column is skewed
```

```python
bank3 = bank1.copy()
```

```python
# Outlier treatment for Age column
Q1 = bank3['Age'].quantile(0.25)
Q3 = bank['Age'].quantile(0.75)
IQR = Q3 - Q1
UL = Q3 + 1.5*IQR
LL = Q1 - 1.5*IQR
```

```python
((bank3['Age'] > UL) | (bank3['Age'] < LL)).sum()
```

```python
(359/bank3.shape[0])*100
```

```python
df = bank3[(bank3['Age'] < UL) & (bank3['Age'] > LL)]
```

```python
# Outlier treatment for Credit score column
upper_limit = df['CreditScore'].mean() + 3 * df['CreditScore'].std()
lower_limit = df['CreditScore'].mean() - 3 * df['CreditScore'].std()
```

```python
df_filtered = df[(df['CreditScore'] >=lower_limit) & (df['CreditScore'] <= upper_limit)]
```

```python
df_filtered.head()
```

```python
bank_4 = df_filtered.drop(columns = 'Tenure_y')
```
